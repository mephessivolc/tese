```python
n=5
```

```python
def create_graph(n):
    m = numpy.triu(numpy.random.randint(1,10, size=(n,n)))
    
    for i in range(n):
        m[i,i] = 0
    G = nx.from_numpy_array(m)
    
    return G

G = create_graph(n)

plt.figure(figsize=(8, 6)) # Define o tamanho da figura
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=1000)
plt.show() # Exibe o gráfico
```

```python
sol = solve_tsp_mtz_ilp(G, start=0, msg=True)
print("Tour:", sol.tour)
print("Custo:", sol.cost)
print("u (posição):", sol.u)
```
