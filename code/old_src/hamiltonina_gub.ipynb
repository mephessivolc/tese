# junto

```python
def calcular_hamiltonian(d_ij, alpha=5, beta=7, gamma=9, dim=3):
    """
    Calcula o Hamiltoniano final como uma combinação de vários termos.
    
    Parameters
    ----------
    d_ij : numpy.ndarray
        Matriz de adjacência do grafo (N x N)
    alpha : float
        Coeficiente para o termo H1
    beta : float
        Coeficiente para o termo H2
    gamma : float
        Coeficiente para o termo H3
    dim : int
        Dimensão do espaço de Hilbert para cada modo quântico
    
    Returns
    -------
    numpy.ndarray
        Matriz do Hamiltoniano final
    """
    
    # Obter número de vértices da matriz
    N = d_ij.shape[0]
    Numero_qmodes = N * (N - 1)
    
    # Criar lista de operadores de destruição
    a_lista = [qt.tensor(*[qt.qeye(dim) if j != i else qt.destroy(dim) 
                           for j in range(Numero_qmodes)]) 
               for i in range(Numero_qmodes)]
    
    # Criar lista de operadores de número
    n_lista = [a.dag() * a for a in a_lista]
    
    # Termo C - interação entre modos
    C = [d_ij[i, j] * n_lista[i] * n_lista[j] 
         for i in range(N) for j in range(N) if i != j]
    C = sum(C)
    
    # Termo H1 - potencial no-lineal
    n_linha = [2 * n_lista[i] - 1 for i in range(len(n_lista))]
    H1 = [alpha * (n_linha[i]**2 - 1)**2 for i in range(len(n_linha))]
    H1 = sum(H1)
    
    # Termo H2
    H2 = []
    for j in range(N):
        H_tempo = 0
        for i in range(N):
            if i != j:
                H_tempo += n_lista[i]
        H2.append(beta * (H_tempo - 1)**2)
    H2 = sum(H2)
    
    # Termo H3
    H3 = []
    for i in range(N):
        H_tempo = 0
        for j in range(N):
            if i != j:
                H_tempo += n_lista[i]
        H3.append(gamma * (H_tempo - 1)**2)
    H3 = sum(H3)
    
    # Termo p - energia cinética
    p = [( 1j * np.sqrt(1/2) * (a.dag() - a) )**2 / 2 for a in a_lista]
    p = sum(p)
    
    # Hamiltoniano final
    H_final = C + H1 + H2 + H3 + p
    
    # Converter para matriz numpy
    return H_final

N = 3
dim = 3

Numero_qmodes = N*(N-1)
print("Numero de subsistema ou qmodes:", Numero_qmodes)

gerador = graphs.GraphBuilder(n=N)
d_ij = gerador.matrix

# Com seus dados atuais
H_final_matriz = calcular_hamiltonian(
    d_ij=d_ij, 
    alpha=100, 
    beta=100, 
    gamma=100, 
    dim=5
)
print(H_final_matriz)
# H_final_matriz é agora um array numpy que pode ser usado em cálculos normais
print(H_final_matriz.shape)  # Ver dimensão da matriz

# for i in range(dim):
#     for j in range(dim):
#         state  = qt.tensor(qt.basis(dim, i),qt.basis(dim, j))
#         print(qt.expect(H_final_matriz, state))
```

```python
energia = []

N= 3
Numero_qmodes = N*(N-1)
dim = 5  # dimensão de cada subsistema

cont = 0
for indices in itertools.product(range(dim), repeat=Numero_qmodes): # gera todas as combinações possíveis de Numero_qmodes índices de 0 a dim-1.
    state = qt.tensor(*[qt.basis(dim, i) for i in indices])
    energia.append(qt.expect(H_final_matriz, state))

    cont += 1
    if cont % 10 == 0:
        print(f"{cont}: {qt.expect(H_final_matriz, state)}")

plt.plot(energia)
plt.show()
```

# Força bruta

```python
from itertools import permutations

def tsp_bruteforce(dist):
    n = len(dist)
    vertices = list(range(1, n))   # fixa o vértice 0 como início
    best_cost = float('inf')
    best_path = None

    for perm in permutations(vertices):
        path = (0,) + perm + (0,)
        cost = 0

        for i in range(len(path) - 1):
            cost += dist[path[i]][path[i + 1]]

        if cost < best_cost:
            best_cost = cost
            best_path = path

    return best_cost, best_path
```

```python
cost, path = tsp_bruteforce(d_ij)
print("Melhor custo:", cost)
print("Melhor rota:", path)
```
